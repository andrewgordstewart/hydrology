# Crop images

```python
# Create functions which return the sky region of a particular set of images

# Works in the Atha 2015/2016 and 2017(a) photos
def getSkyAthaA(img):
    return img.crop((650,0,1300,450))

# Works in the Atha 2017(b) and 2018 photos
def getSkyAthaB(img):
    return img.crop((1800,0,2600,450))

def getSkyPeyto15(img):
    return img.crop((0,0,3200,200))

def getSkyPeyto15B(img):
    return img.crop((1280,0,2700,250))

def getSkyPeyto16(img):
    return img.crop((200,0,3200,360))

def getSkyPeyto18(img):
    return img.crop((1000,0,3200,610))

def getSkyPeyto18B(img):
    return img.crop((580,0,3200,1300))

### Check the boxes with the following statements
# athaA = Image.open('../data/photos/original/Atha/2015/WSCT1155.JPG')
# athaB = Image.open('../data/photos/original/Atha/2017/b/WSCT3381.JPG')
# plt.imshow(athaA, cmap = 'gray', interpolation = 'bicubic')
# plt.imshow(getSkyAthaA(athaA))
# plt.imshow(getSkyAthaB(athaB))
```

```python
# Return a sorted list of all subdirectories of path that contain a JPG image
def getImageFolders(path):
    folders = set()
    for root, dirs, files in os.walk(path):
        for file in files:
            if file.endswith('.JPG'):
                folder = os.path.join(root)
                folders.add(folder)
    return list(sorted(folders))


# Creates $DATA_DIRECTORY/photos/sky/$DESCRIPTOR for each $DESCRIPTOR,
# where a $DESCRIPTOR is the $POSTFIX of a terminal subdirectory of
# $DATA_DIRECTORY/photos/original/$POSTFIX
def createSkyDataDirectories(folders):
    prefix = os.path.commonpath(folders)
    for f in folders:
        os.makedirs(
            os.path.join(dataDirectory, 'photos', 'sky', f[len(prefix) + 1:]),
            exist_ok=True
        )

athaAFolders = [
    '../data/photos/original/Atha/2015',
    '../data/photos/original/Atha/2016',
    '../data/photos/original/Atha/2017/a'
]
athaBFolders = [
    '../data/photos/original/Atha/2017/b',
    '../data/photos/original/Atha/2018' 
]
peyto2015Folders = [
    '../data/photos/original/Peyto/2015'
]
peyto2015BFolders = [
    '../data/photos/original/Peyto/2015/b'
]
peyto2016Folders = [
    '../data/photos/original/Peyto/2016'
]
peyto2018Folders = [
    '../data/photos/original/Peyto/2018/b'
]
peyto2018bFolders = [
    '../data/photos/original/Peyto/2018/'
]
# Crops all images around photos found in folders, writing images to $DATA_DIRECTORY/sky/...
def cropImages(folders):
    for folder in folders:
        for root, dirs, files in os.walk(folder):
            for file in files:
                if file.endswith('.JPG'):
                    folder = os.path.join(root)
                    filename = os.path.join(folder, file)
                    
                    image = Image.open(filename)
                    if (folder in athaAFolders):
                        cropped = getSkyAthaA(image)
                    
                    elif (folder in athaBFolders):
                        cropped = getSkyAthaB(image)
                    
                    elif (folder in peyto2015Folders):
                        cropped = getSkyPeyto15(image)
                    
                    elif (folder in peyto2015BFolders):
                        cropped = getSkyPeyto15B(image)
                    
                    elif (folder in peyto2016Folders):
                        cropped = getSkyPeyto16(image)
                    
                    elif (folder in peyto2018Folders):
                        cropped = getSkyPeyto18(image)
                    
                    elif (folder in peyto2018bFolders):
                        cropped = getSkyPeyto18B(image)
                    
                    else:
#                         raise Exception('Invalid folder', folder)
                        continue # Do nothing for now
                    cropped.save(filename.replace('original', 'sky'))
```

## Preprocessing

```python
dataDirectory = '../data/'
folders = getImageFolders(dataDirectory)
createSkyDataDirectories(folders)
cropImages(folders)
```
